```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sen(x, b):
  return np.sin(2*np.pi*x-b)

x=np.linspace(0, 2, 500)
b=np.arange(0, 2, 0.25)
```

```python
#CREO LA LISTA DE COLORES
#algunas paletas: "flare", "crest", "magma", "viridis", "rocket_r", "cubehelix", "Blues","YlOrBr" 
# generating an arbitrary cubehelix palette:"ch:s=0,r=.5" s es un numero entre 0 y 3. r entre -1 y 1
pal = sns.color_palette("ch:s=2.5,r=0", len(b))  
 
plt.figure(figsize=(10,6)), plt.grid(True)
for i in range(len(b)):
  plt.plot(x, sen(x, b[i]), color=pal[i], linewidth=3)
plt.ylabel("Amplitud", fontsize=20)
```
